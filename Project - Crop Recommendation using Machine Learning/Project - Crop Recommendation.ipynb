```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
data = {
    'soil_ph': [6.5, 7.0, 5.5, 6.0, 8.0, 7.2, 6.8, 5.0, 6.1, 7.5],
    'soil_moisture': [20, 15, 30, 25, 10, 20, 18, 40, 28, 12],
    'temperature': [22, 25, 30, 20, 15, 28, 24, 32, 26, 18],
    'humidity': [70, 65, 80, 75, 60, 68, 72, 85, 78, 63],
    'crop_type': ['Wheat', 'Rice', 'Rice', 'Wheat', 'Corn', 'Corn', 'Wheat', 'Rice', 'Wheat', 'Corn']
}
df = pd.DataFrame(data)
```

```python
X = df[['soil_ph', 'soil_moisture', 'temperature', 'humidity']]
y = df['crop_type']
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model 1: Decision Tree Classifier
model_tree = DecisionTreeClassifier()
model_tree.fit(X_train, y_train)
y_pred_tree = model_tree.predict(X_test)

# Model 2: K-Nearest Neighbors Classifier
model_knn = KNeighborsClassifier(n_neighbors=3)
model_knn.fit(X_train, y_train)
y_pred_knn = model_knn.predict(X_test)
```

```python
# Calculate accuracy scores for both models
accuracy_tree = accuracy_score(y_test, y_pred_tree)
accuracy_knn = accuracy_score(y_test, y_pred_knn)

print(f"Decision Tree Accuracy: {accuracy_tree * 100:.2f}%")
print(f"K-Nearest Neighbors Accuracy: {accuracy_knn * 100:.2f}%")
```
